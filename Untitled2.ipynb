```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('data.csv')

data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
X = data['ACCEX'].values[:,np.newaxis]

y = data['LIGHT'].values

model = LinearRegression()

model.fit(X, y)
```

```python
plt.scatter(X, y,color='r')

plt.plot(X, model.predict(X),color='k')
plt.xlabel('ACCEX')
plt.ylabel('LIGHT')
plt.show()
```
